```python
from ultralytics import YOLO
import os

from google.colab import drive
drive.mount('/content/drive')

DATA = '/content/drive/MyDrive/high_jump.v9i.yolov8/data.yaml'
SAVE_DRIVE = '/content/drive/MyDrive/high_jump_models'
PROJECT = 'high_jump_runs'

os.makedirs(SAVE_DRIVE, exist_ok=True)
```

```python
def run_experiment(model_name, exp_name, epochs=100, imgsz=640, batch=16,
                   lr0=None, lrf=None,
                   mosaic=None, mixup=None, copy_paste=None,
                   fliplr=None, degrees=None, scale=None, shear=None,
                   hsv_h=None, hsv_s=None, hsv_v=None,
                   freeze=None, patience=None):

    print(f"\nRunning experiment: {exp_name}\n")

    model = YOLO(model_name)

    train_args = dict(
        data=DATA,
        epochs=epochs,
        imgsz=imgsz,
        batch=batch,
        project=PROJECT,
        name=exp_name,
        deterministic=True,
        seed=0,
    )

    optional_params = {
        "lr0": lr0, "lrf": lrf,
        "mosaic": mosaic, "mixup": mixup, "copy_paste": copy_paste,
        "fliplr": fliplr, "degrees": degrees, "scale": scale, "shear": shear,
        "hsv_h": hsv_h, "hsv_s": hsv_s, "hsv_v": hsv_v,
        "freeze": freeze, "patience": patience
    }

    for k, v in optional_params.items():
        if v is not None:
            train_args[k] = v

    model.train(**train_args)

    best_path = f"/content/{PROJECT}/{exp_name}/weights/best.pt"
    best_model = YOLO(best_path)

    print("\nEvaluating on TEST set...\n")
    metrics = best_model.val(data=DATA, split='test')

    save_to = f"{SAVE_DRIVE}/{exp_name}_best.pt"
    !cp {best_path} {save_to}
    print(f"\nSaved model to {save_to}\n")

    return metrics
```

### 1A - YOLOv8n

```python
run_experiment(
    model_name='yolov8n.pt',
    exp_name='exp_v8n_640'
)
```

### 1B - YOLOv8s (baseline)

```python
run_experiment(
    model_name='yolov8s.pt',
    exp_name='exp_v8s_640'
)
```

### 1C - YOLOv8m

```python
run_experiment(
    model_name='yolov8m.pt',
    exp_name='exp_v8m_640',
    batch=8
)
```

### 2A - Strong augmentation

```python
run_experiment(
    model_name='yolov8s.pt',
    exp_name='exp_v8s_strong_aug',
    mosaic=0.0,
    mixup=0.0,
    copy_paste=0.0,
    fliplr=0.5,
    scale=0.2,
    hsv_h=0.005,
    hsv_s=0.2,
    hsv_v=0.2,
)
```
